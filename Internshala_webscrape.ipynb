**BeautifulSoup and Request Webscraping**

```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
```

Getting the URL



```python
url = 'https://internshala.com/fresher-jobs/data-science-jobs-in-gurgaon/'
```

```python
page = requests.get(url)
```

Parsing through BeautifulSoup

```python
soup = BeautifulSoup(page.text, 'lxml')
```

```python
def page_code():
  return print(soup.prettify())
```

Preliminary Check

```python
company_names = []
for i in soup.find_all('p', class_='company-name'):
    print(i.text.strip())
```

Creating Lists

```python
company = soup.find_all('div', class_ = 'internship_meta')
```

```python
name = []
job_title = []
stipend = []
skills = []
location = []

for i in company:
  try:
    name.append(i.find('p', class_ = 'company-name').text.strip())
  except AttributeError:
    name.append(None) # or some other placeholder

  try:
    job_title.append(i.find('a', class_ ='job-title-href').text.strip())
  except AttributeError:
    job_title.append(None) # or some other placeholder
  try:
    stipend.append(i.find('span', class_ ='desktop').text.strip())
  except AttributeError:
    stipend.append(None)
  try:
    skills.append(i.find('div', class_='job_skills').text.strip())
  except AttributeError:
    skills.append(None)
  try:
    location.append(i.find('p', class_ ='row-1-item locations').text.strip())
  except AttributeError:
    location.append(None) # or some other placeholder
```

Converting to Dataframe

```python
d={'name' :name, 'job_title':job_title, 'stipend':stipend, 'skills':skills, 'location':location}
pd.DataFrame(d)
```
